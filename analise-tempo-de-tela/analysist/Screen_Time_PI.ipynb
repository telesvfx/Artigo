```python
# ðŸ” 4. ConversÃµes Ãºteis
# Criando uma coluna numÃ©rica para tempo mÃ©dio de tela
def converter_tempo(t):
    if t == '4â€“6':
        return 5
    elif t == '6â€“8':
        return 7
    elif t == '8-10':
        return 9
    elif t == 'More than 10':
        return 11
    else:
        return None

df['ScreenTimeNum'] = df['Average Screen Time'].apply(converter_tempo)
```

```python
# ðŸŽ¯ 5. CorrelaÃ§Ãµes e agrupamentos
# Tempo mÃ©dio por dispositivo
media_tempo_por_dispositivo = df.groupby('Device')['ScreenTimeNum'].mean().reset_index()
sns.barplot(data=media_tempo_por_dispositivo, x='Device', y='ScreenTimeNum', palette='coolwarm')
plt.title('Tempo MÃ©dio de Tela por Dispositivo')
plt.ylabel('Horas Aproximadas')
plt.show()
```

```python
import pandas as pd

# FunÃ§Ã£o robusta
def extrair_idades(faixa):
    faixa = faixa.strip().replace('â€“', '-').replace('âˆ’', '-')  # padroniza traÃ§os e remove espaÃ§os
    faixa_lower = faixa.lower()

    if 'more than' in faixa_lower:
        idade_min = int(faixa_lower.split()[-1])
        idade_max = idade_min + 2  # estimativa
    elif 'less than' in faixa_lower:
        idade_max = int(faixa_lower.split()[-1])
        idade_min = max(idade_max - 2, 0)  # nÃ£o deixar negativo
    else:
        idade_min, idade_max = map(int, faixa.split('-'))

    return pd.Series([idade_min, idade_max])

# Aplicar a funÃ§Ã£o
df[['screen_min', 'screen_max']] = df['Average Screen Time'].apply(extrair_idades)

# Calcular a mÃ©dia da faixa
df['screen_avg'] = (df['screen_min'] + df['screen_max']) / 2

# MÃ©dia geral
media_geral = df['screen_avg'].mean()
print(media_geral)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Corrigir valores ausentes
df.fillna("NÃ£o informado", inplace=True)

# Tabela cruzada entre tempo de tela e produtividade
cross_tab = pd.crosstab(df["Average Screen Time"], df["Productivity"], normalize='index') * 100
cross_tab = cross_tab.round(2)

# Reordenar colunas se necessÃ¡rio
cols_order = [
    "Extremely productive, i efficiently complete my tasks",
    "Moderately productive",
    "Unproductive, i might not have completed the task and got carried away"
]
cross_tab = cross_tab[cols_order]

# Printar a tabela formatada como no exemplo
print("\nðŸ§  Como interpretar os dados:")
print("Tempo de Tela (h/dia)\tExtremamente Produtivo\tModeradamente Produtivo\tPouco Produtivo")
for idx, row in cross_tab.iterrows():
    print(f"{idx}\t{row[0]:.2f}%\t\t\t{row[1]:.2f}%\t\t\t{row[2]:.2f}%")

# GrÃ¡fico de barras empilhadas
cross_tab.plot(kind='bar', stacked=True, colormap='viridis', figsize=(10, 6))
plt.title("Produtividade Relativa por Tempo MÃ©dio de Tela")
plt.xlabel("Tempo MÃ©dio de Tela (h/dia)")
plt.ylabel("DistribuiÃ§Ã£o de Produtividade (%)")
plt.legend(title="Produtividade")
plt.tight_layout()
plt.show()
```
